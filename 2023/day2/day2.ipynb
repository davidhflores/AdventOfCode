```python
import pandas as pd
import numpy as np
```

# Part 1

```python
filename='/content/drive/MyDrive/Colab Notebooks/advent_of_code_2023/day2/input2.txt'

with open(filename,'r+') as file1:
  lines_text= file1.readlines()

# see sample
lines_text[:5]
```

```python
config = {
    'red':12,
    'green':13,
    'blue':14
}
```

```python
i=0
df_output = pd.DataFrame()
for line in lines_text:
  i+=1
  line = line.replace('\n','')
  game_data={'game':i}
  games=line.split(': ')[1]
  hands=games.split('; ')
  h=0
  for hand in hands:
    h+=1
    game_data[f'hand_{h}']={}
    cubes=hand.split(', ')
    for cube in cubes:
      num_cube = cube.split(' ')
      game_data[f'hand_{h}'][num_cube[1]]=int(num_cube[0])
  df_dict = pd.DataFrame([game_data])
  df_output = pd.concat([df_output,df_dict], ignore_index=True)
```

```python
df_output = df_output.fillna('')
df_output
```

```python
# Rules: red 12,green 13,blue 14
valid_games = []
for i,row in df_output.iterrows():
  game = []
  for col in range(1,len(df_output.columns)):
    if row[f'hand_{col}']!='':
      hands=[]
      for k,v in row[f'hand_{col}'].items():
        if v <= config[k]:
          hands.append(True)
        else:
          hands.append(False)
      if all(x for x in hands):
        game.append(True)
      else:
        game.append(False)
  if all(x for x in game):
    valid_games.append(row['game'])
```

```python
sum(valid_games)
```

# Part 2

```python
games_min=pd.DataFrame()
for i,row in df_output.iterrows():
  min_color={}
  for col in range(1,len(df_output.columns)):
    if row[f'hand_{col}']!='':
      for k,v in row[f'hand_{col}'].items():
        if k not in min_color.keys():
          min_color[k] = v
        elif v > min_color[k]:
          min_color[k] = v
  data = pd.DataFrame({'game':row['game'],'min_color':[min_color]})
  games_min = pd.concat([games_min,data],ignore_index=True)
```

```python
games_min.head()
```

```python
game_power=[]
for i,row in games_min.iterrows():
  game_power.append(
      np.prod(
          list(row['min_color'].values())
      )
  )

sum(game_power)
```
